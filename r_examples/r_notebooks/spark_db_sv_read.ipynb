```r
library(SparkR)
library(ibmWatsonStudioLib)
wslib <- access_project_or_space()
library(itcutils)

# NOTE:
#  A limit of 5000 rows has been applied to the request to enable sample previewing.
#  Adjust the display message as needed by editing the following lines:
library(IRdisplay)
display_html("A row limit of 5000 has been applied to the query to enable sample previewing. If the data set is larger, only the first 5000 rows will be loaded.")
#  Edit select_statement to change or disable the row limit.
#
nb_data_request = list(
    "connection_name" = "Db2oltp-1tb",
    "interaction_properties" = list(
        "select_statement" = "SELECT * FROM \"ANZ\".\"BANK_ACCOUNTS\" FETCH FIRST 5000 ROWS ONLY"
    )
)

invisible(sparkR.session(appName = "SparkSession R", sparkConfig = list(spark.sql.execution.arrow.sparkr.enabled = "true")))
df_data_1 <- read.df("",
    "com.ibm.connect.spark.flight",
    flight.location = itcutils::get_flight_service_url(),
    flight.command = itcutils::get_flight_cmd(nb_data_request,wslib),
    flight.authToken = itcutils::get_bearer_token()
)
head(df_data_1)

```

```r
# @hidden_cell

library(ibmWatsonStudioLib)
wslib <- access_project_or_space()

Db2oltp_1tb_metadata = wslib$get_connection("Db2oltp-1tb")

library(SparkR)
invisible(sparkR.session(appName = "SparkSession R"))

Db2oltp_1tb_url <- paste("jdbc:db2://",
    Db2oltp_1tb_metadata[][["host"]],
    ":", Db2oltp_1tb_metadata[][["port"]],
    "/", Db2oltp_1tb_metadata[][["database"]],
    ":", "sslConnection=true;",
    sep=""
)

df_data_2 <- read.jdbc(url = Db2oltp_1tb_url,
    table = "\"ANZ\".\"BANK_ACCOUNTS\"",
    user = Db2oltp_1tb_metadata[][["username"]],
    password = Db2oltp_1tb_metadata[][["password"]]
)
head(df_data_2)

```
